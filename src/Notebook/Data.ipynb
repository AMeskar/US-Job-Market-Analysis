```python
def execution(words, list_val):

    arr = []

    for val in list_val:

        if val.lower() in words.lower():

            arr.append(val)
    
    if arr == [] : arr.append('No Requirements are Specified')

    return arr
```

```python
df_new['skills'] = df_new['description'].apply(lambda x: execution(x, skills))
df_new['education'] = df_new['description'].apply(lambda x: execution(x, education))
df_new['concepts'] = df_new['description'].apply(lambda x: execution(x, concepts))
df_new['majors'] = df_new['description'].apply(lambda x: execution(x, majors))
```
